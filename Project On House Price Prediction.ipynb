## Train and Test Splitting

```python
#from sklearn.model_selection import train_test_split
#train_set, test_set  = train_test_split(housing, test_size=0.2, random_state=42)
#print(f"Rows in train set: {len(train_set)}\nRows in test set: {len(test_set)}\n")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['waterfront']):
    train_set = housing.loc[train_index]
    test_set = housing.loc[test_index]
print(f"Rows in train set: {len(train_set)}\nRows in test set: {len(test_set)}\n")
```

## Training The Model

```python
import numpy as np
from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

## Linear Regression

```python
model_1=linear_model.LinearRegression()
x_train1 = train_set.drop("price", axis=1)
y_train1 = train_set["price"].copy()           
model_1.fit(x_train1,y_train1)
```

## Decission Tree

```python
model_2=DecisionTreeRegressor()
x_train2 = train_set.drop("price", axis=1)
y_train2 = train_set["price"].copy()           
model_2.fit(x_train2,y_train2)
```

## Random Forest

```python
model_3=RandomForestRegressor()
x_train3 = train_set.drop("price", axis=1)
y_train3 = train_set["price"].copy()           
model_3.fit(x_train3,y_train3)
```

## Predicting the Price using each model

```python
x_test1 = test_set.drop("price", axis=1)
y_test1 = test_set["price"].copy()
pred_1=model_1.predict(x_test1)

x_test2 = test_set.drop("price", axis=1)
y_test2 = test_set["price"].copy()
pred_2=model_2.predict(x_test2)

x_test3 = test_set.drop("price", axis=1)
y_test3 = test_set["price"].copy()
pred_3=model_3.predict(x_test3)
```

## Testing the Model

```python
print('linear model :-')
mean_squared_error_1=mean_squared_error(y_test1,pred_1)
print('Sqaured mean error', round(np.sqrt(mean_squared_error_1),2))
print('R squared training',round(model_1.score(x_train1,y_train1),3))
print('R sqaured testing',round(model_1.score(x_test1,y_test1),3) )

print('Decission Tree model :-')
mean_squared_error_2=mean_squared_error(y_test2,pred_2)
print('Sqaured mean error', round(np.sqrt(mean_squared_error_2),2))
print('R squared training',round(model_2.score(x_train2,y_train2),3))
print('R sqaured testing',round(model_2.score(x_test2,y_test2),3) )

print('Random Forest model :-')
mean_squared_error_3=mean_squared_error(y_test3,pred_3)
print('Sqaured mean error', round(np.sqrt(mean_squared_error_3),2))
print('R squared training',round(model_3.score(x_train3,y_train3),3))
print('R sqaured testing',round(model_3.score(x_test3,y_test3),3) )
```

## Cross Validation Method

```python
from sklearn.model_selection import cross_val_score
scores1 = cross_val_score(model_1, x_train1, y_train1, scoring="neg_mean_squared_error", cv=10)
rmse_scores1 = np.sqrt(-scores1)

scores2 = cross_val_score(model_2, x_train2, y_train2, scoring="neg_mean_squared_error", cv=10)
rmse_scores2 = np.sqrt(-scores2)

scores3 = cross_val_score(model_3, x_train3, y_train3, scoring="neg_mean_squared_error", cv=10)
rmse_scores3 = np.sqrt(-scores3)
```
